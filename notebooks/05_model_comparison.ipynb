# Model Comparison

This notebook compares the performance of Logistic Regression and Random Forest models for fraud detection.

The evaluation focuses on:
- Precision
- Recall
- F1-score


```python
import pandas as pd
import matplotlib.pyplot as plt
import os

#Load classification reports
df_lr = pd.read_csv("../results/classification_report_logistic.csv", index_col=0)
df_rf = pd.read_csv("../results/classification_ Random_Forest_Baseline.csv", index_col=0)
```

```python
#Extract fraud class metrics
lr_fraud = df_lr.loc["1"]
rf_fraud = df_rf.loc["1"]
```

```python
#Results
df_results = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest (Baseline)"],
    "Precision (Fraud)": [lr_fraud["precision"], rf_fraud["precision"]],
    "Recall (Fraud)": [lr_fraud["recall"], rf_fraud["recall"]],
    "F1-score (Fraud)": [lr_fraud["f1-score"], rf_fraud["f1-score"]]
})

#Round for readability - 2 places
df_results = df_results.round(2)

df_results
```

```python
#Bar plot

import matplotlib.pyplot as plt
import os

df_results.set_index("Model").plot(kind="bar", figsize=(8,5))

plt.title("Model Comparison (Fraud Class)")
plt.ylabel("Score")
plt.xticks(rotation=0)

os.makedirs("../reports/figures", exist_ok=True)
plt.savefig("../reports/figures/model_comparison.png", bbox_inches='tight')

plt.show()
```

# Key Findings

- Random Forest achieved slightly higher precision and F1-score
- Logistic Regresion score significantly higher Recall.
