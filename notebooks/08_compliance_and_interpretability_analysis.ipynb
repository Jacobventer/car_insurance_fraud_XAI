# Compliance and Interpretability Analysis

This section evaluates the transparency and interpretability of the machine learning models used for fraud detection.

It further examines how explainable AI techniques support regulatory requirements such as the General Data Protection Regulation (GDPR).

# Model Transparency

Logistic Regression is inherently interpretable, as it provides clear relationships between input features and predictions.

In contrast, Random Forest is a more complex model and operates as a black-box, making it difficult to interpret without additional tools.

This highlights the need for explainable AI techniques to improve transparency in complex models.

# Explainability Analysis

SHAP and LIME were applied to improve interpretability.

SHAP provided both global and local explanations, identifying key features such as VehicleCategory, Fault, and BasePolicy as major contributors to fraud detection.

LIME complemented this by explaining individual predictions, showing how specific feature values influenced model outcomes.

# Performance vs Interpretability Trade-off

The results show that Logistic Regression achieved performance comparable to Random Forest, despite being more simple.

This suggests that simpler models can be effective in fraud detection.

# Compliance with GDPR

The General Data Protection Regulation (GDPR), particularly Article 22, requires transparency in automated decision-making.

Explainable AI techniques such as SHAP and LIME help fulfil these requirements by providing insights into how decisions are made.

These methods support:
- Transparency
- Accountability
- The ability to justify decisions

This is particularly important in fraud detection, where decisions may impact individuals financially.

# Risks and Limitations

Several limitations were identified:

- The dataset is imbalanced, which affects model performance.
- The dataset is publicly available and may not reflect real-world complexity.
- Explainability methods provide approximations and may not fully capture model behaviour.
- Potential bias in the data may affect fairness and decision-making.

These factors should be considered when interpreting results.

# Conclusion

Explainable AI techniques significantly improve the transparency and interpretability of machine learning models in fraud detection.

The combination of SHAP and LIME enables both global and local understanding of model behaviour.

This supports regulatory compliance and increases trust in automated decision-making systems.
